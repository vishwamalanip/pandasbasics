# Merging and Concat in Pandas

# Merge

```python
import pandas as pd
```

```python
#in merging it should be something exaclty common
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,4,5],"V":[11,12,13,14,15]})

pd.merge(d,s,on="A") # on="key name" which is common in both dataframe
```

```python
#in merging it should be something exaclty common, if not common than it will merge only upto the common data  #in this one {"A":[6,7] is not same
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"V":[11,12,13,14,15]})

pd.merge(d,s,on="A") # on="key name" which is common in both dataframe
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"V":[11,12,13,14,15]})

pd.merge(d,s,how= 'left')  
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"V":[11,12,13,14,15]})

pd.merge(d,s,how= 'right') #how = 'right' it will merge all values which are common but which is not common it will put Nans there in right column
```

```python
#it will put nan right and left both to those which are not matching
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"V":[11,12,13,14,15]})

pd.merge(d,s,how= 'outer')#it will show only common data #works like union
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"V":[11,12,13,14,15]})

pd.merge(d,s,how= 'outer',indicator= True) # to check which data is present and which is not
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"S":[11,12,13,14,15]})

pd.merge(d,s) #this will not show anything
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"S":[11,12,13,14,15]})

pd.merge(d,s,left_index= True, right_index= True) #left_index and right_index will show all the data 
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"S":[11,12,13,14,15]})

pd.merge(d,s,left_index= True, right_index= True, suffixes = ("name","email")) #to give name to column
```

# Concat

```python
s=pd.Series([1,2,3,4])
s1=pd.Series([11,21,31,41])

s
s1
```

```python
s=pd.Series([1,2,3,4])
s1=pd.Series([11,21,31,41])
pd.concat([s,s1]) # to merge both series
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"S":[11,12,13,14,15]})
pd.concat([d,s])  # to merge both dataframe
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"S":[11,12,13,14,15]})
pd.concat([d,s],axis=1)  # if axis=1 it will merge according to columns ,but if axis=0 it will merge according to rows just like above output
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2],"S":[11,12]})
pd.concat([d,s],axis=1) 
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2],"S":[11,12]})
pd.concat([d,s],axis=1,join="outer") #if outer join it will show same output it will show missing data aswell
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2],"S":[11,12]})
pd.concat([d,s],axis=1,join="inner") #if inner join it will show only those data which dont has Nans #if will not show missing data
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"C":[11,12,13,14,15]})
pd.concat([d,s],axis=1,keys=['Python','C++'])  #it will give name to dataframe 
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5],"S":[6,7,8,9,10]})
s = pd.DataFrame({"A":[1,2,3,6,7],"C":[11,12,13,14,15]})
pd.concat([d,s],axis=0,keys=['Python','C++'])  #it will give name to dataframe 
```

```python
d = pd.DataFrame({"A":[1,2,3,4,5]})
s = pd.DataFrame({"A":[1,2,3,6,7],"C":[11,12,13,14,15]})
pd.concat([d,s])  #it will merge like this aswell
```

```python
d = pd.DataFrame({"B":[1,2,3,4,5]})
s = pd.DataFrame({"A":[1,2,3,6,7],"C":[11,12,13,14,15]})
pd.concat([d,s])  #it will merge like this aswell
```
