```python
#droping missing data which has value NaN ofcolumn
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
c= csv_1.head()
c
c.dropna(how="all")
```

```python
#to remove values Nan values of particular column
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.dropna(subset=["hotel"]) #to remove values of which rows has Nan values for hotel
```

```python
#inplace #ek hi column ke under multiple NaN values delete
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.dropna(inplace=True) 
csv_1
```

```python
#ek hi column ke under single NaN values delete that only
'''
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.dropna(thresh=1) 
csv_1
'''
#ek hi column ke under 3 NaN values delete that only
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.dropna(thresh=3) 
csv_1
```

```python
#change NaN value to something in whole document
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
c = csv_1.fillna("python") 
c.head()
```

```python
#change NaN value to something to particular row and data
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
c = csv_1.fillna({"agent":"python","company":'c++',"required_car_parking_spaces":"java"}) 
c.head()
```
