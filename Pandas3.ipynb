# Interpolate

```python
import pandas as pd
```

```python
#interpolate only works with numbers not alphabets if you only use interpolate() directly
csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")

csv_1.interpolate() #to fill missing values automatically check for lead_time ,arrival_date_yea
```

# Methods of interpolate
#method:{'linear','time','index','values','nearest' ,'zero','slinear','cubic','krogh','polynomial','spline','pchip','akima'}

```python

# interpolate only works with numbers not alphabets 
# it has diff methods 

csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.interpolate(method='linear') #data will fill linearly
```

```python

csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.interpolate(method='linear',axis=0)  #if axis=0 it will fill column by column 
                                            #if axis=1  it will fill row by row but during row by row each value has to be same data type
```

```python

csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.interpolate(limit=3)  #in each column it will fill only up to 3 Nans after that all Nans would be same
```

```python

csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.interpolate(limit_direction="forward", limit=3) #limit_direction can change by forward(from up it will change 3 nans from where it starts)/backward(from below it will change 3 nans from where it starts)/both (it will change in both forward and backward direction)
```

```python

csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.interpolate(limit_area='inside') #it will fill all the blank data including Nans ,but if limit_area='outside'it will only change blankdata except Nans
```

```python

csv_1 = pd.read_csv("/Users/vishwamalani/Desktop/python/hotel_booking.csv")
csv_1.interpolate(limit_direction="forward", limit=3 ,inplace=True) #it will change in original file ,but if inplace=False it will make only copy of data
```
