# Pivot Table and Melt Function in Pandas to reshape and file

# Pivot()

```python
import pandas as pd
```

```python
var = pd.DataFrame({'days':[1,2,3,4,5,6],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var
```

```python
var = pd.DataFrame({'days':[1,2,3,4,5,6],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var.pivot(index="days",columns="name") #it will show everything accoring to names where index is going to be days
```

```python
var = pd.DataFrame({'days':[1,2,3,4,5,6],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var.pivot(index="days",columns="name",values='eng') #to only get eng data not maths
```

```python
var = pd.DataFrame({'days':[1,1,1,1,2,2],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var.pivot_table(index='name',columns='days',aggfunc='mean')  #to get mean or max of any data 
```

```python
var = pd.DataFrame({'days':[1,1,1,1,2,2],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var.pivot_table(index='name',columns='days',aggfunc='sum') #to get sum or max of any data 
```

```python
var = pd.DataFrame({'days':[1,1,1,1,2,2],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var.pivot_table(index='name',columns='days',aggfunc='sum',margins="True") #margin by default gives mean/average
```

```python
var = pd.DataFrame({'days':[1,1,1,1,2,2],'name':['vishwa','sagar','jayveersir','vishwa','sagar','jayveersir'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var.pivot_table(index='name',columns='days',aggfunc='mean',margins="True") #it will give mean
```

# Melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame({'days':[1,2,3,4,5,6],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
var #by default shows horizontally data
```

```python
var = pd.DataFrame({'days':[1,2,3,4,5,6],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
pd.melt(var) #shows vertically reshape data 
```

```python
var = pd.DataFrame({'name':['vishwa','sagar','jayveersir','bhoomi','krutang','shahrukh'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
pd.melt(var,id_vars=['name']) #it will take id name and shows eng and maths marks of them
```

```python
var = pd.DataFrame({'name':['vishwa','sagar','jayveersir','bhoomi','krutang','shahrukh'],'eng':[10,12,14,15,16,12],'maths':[17,18,19,13,14,16]})
pd.melt(var,id_vars=['name'],var_name="subject",value_name='scores')  #to give variable name and value name
```
